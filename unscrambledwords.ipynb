```python
''' based on a hacker rank challenge to unscramble several words the originals of which are
    held in a longer list of words supplied in a file '''

from random import shuffle
from random import choice

def scramble(word):
    ''' returns string with character order randomly shuffled'''
    word = list(word)
    shuffle(word)
    return ''.join(word)

def scramblewords(wordlist, quantity):
    ''' return list of <quantity> scrambled words randomly chosen from <wordlist> '''
    words = []
    words_scrambled = []
    for _ in range(quantity + 1):
        while True:
            word = choice(wordlist)
            if word not in words: break
        words.append(word)
        words_scrambled.append(scramble(word))
    return words_scrambled, words

def read_words_from_file():
    ''' read file of words '''
    with open('/Users/stuartmoore/IDrive-Sync/pythonscratch/wordlist.txt', 'r') as f:
        worddict = {}  # sorted string versions of words (keys), with original words (values)
        wordlist = []  # original words, from which words can be selected randomly for scrambling
        for word in f:
            simpleword = word.strip().lower()
            sortedword = ''.join(sorted(list(simpleword)))
            if not worddict.get(sortedword):
                worddict[sortedword] = simpleword
                wordlist.append(simpleword)
    return worddict, wordlist

def solve_scrambled_word(scrambled, worddict):
    ''' given a scambled word, return original '''
    sortedword = ''.join(sorted(scrambled))
    return worddict.get(sortedword)

def main():

    worddict, wordlist = read_words_from_file()
    words_scrambled, assertions = scramblewords(wordlist, quantity=6)
    
    # add test at end to check testing is working
    words_scrambled.append('FAILEDTEST')
    assertions.append('THIS SHOULD FAIL')
    
    print(f"\n\nWords to unscramble are: {', '.join(words_scrambled)}:\n")    

    for word, assertion in zip(words_scrambled, assertions):
        try:
            found = solve_scrambled_word(word, worddict)
            assert found == assertion
        except AssertionError:
            print(f'*** ERROR **>> {word} does not have a match.')
        except Exception as e:
            print(f'*** ERROR **>>\n{e}')
        else:
            print(f'{word} is {found}.')

if __name__ == "__main__":
    main()
```
