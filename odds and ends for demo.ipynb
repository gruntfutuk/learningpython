```python
''' number guessing game '''
from random import randint

TOP = 100
BOTTOM = 1

# play again loop
while True:
    
    found = False
    tries = 10
    target = randint(BOTTOM, TOP)
    
    # game loop
    while not found and tries > 0:

        # valid input loop
        while True:
            player = input(f'Guess ({BOTTOM} - {TOP})  (return to exit): ')
            if not player:  # player just pressed return
                break
            if player.isnumeric():
                guess = int(player)
                if BOTTOM <= guess <= TOP:
                    break
                print(f'{player} is not within the specific range')
            else:
                print(f'{player} is not numeric.')

        if not player:  # player just pressed return
            break

        found = guess == target

        if not found:
            tries -= 1
            direction = 'lower' if target < guess else 'higher'
            print(f'Sorry, not right. Target is {direction}. Number of tries left: {tries}')

    else:
        if found:
            print(f'Well done. You managed this in {tries} goes.')
        else:
            print('Better luck next time')
    
    if input('Play again? (Y/N)').lower() not in ['y', 'yes', 'yup', 'ok']:
        break
```

```python
''' rock paper scissors Spoke lizards'''

def rock(com_play, user_play, *plays):
    plays = list(plays)
    while plays[0] != com_play:
        shifted = plays.pop(0)
        plays.append(shifted)
    game_state = ['draw', 'You lose', 'You win', 'You lose', 'You win'] \
        if len(plays) == 5 else ['draw', 'You win', 'You lose']
    return game_state[plays.index(user_play)]
```

```python
def stuart_count2(source):
    ''' return count occurrences of last 2 char 
    substring in preceding part of source string '''
    count = 0
    if source[-3:-2]:
        search = source[-2:]
        prev = source[0]
        for ch in source[1:-2]:
            if prev + ch == search:
                count += 1
            prev = ch
    return count
```

```python
def mark_count2(target):
    count = 0
    tail = target[-2:]
    while target != tail:
        count += target[:2] == tail
        target = target[1:]
    return count
```

```python
def str_to_numlst(sample):
    s = sample
    return [num for comma_sep in sample.split(',')
                for before, _, after in [comma_sep.partition('-')]
                for num in range(int(before), int(after or before)+1)
            ]
```

```python
def str_to_numlst_valid(sample):
    result = []
    errmsg = ''
    source = "".join(sample.split())
    for comma_sep in source.split(','):
        before, hyphen, after = comma_sep.partition('-')
        if (not before or not before.isnumeric()) or \
                (hyphen and not after) or \
                (after and not after.isnumeric()):
            errmsg += f'{sample} contains invalid number/range in {comma_sep}' + '\n'
        else:
            start = int(before)
            fini = int(after) if after else start
            if start > fini:
                start, fini = fini, start
            for num in range(start, fini+1):
                result.append(num)
    if errmsg:
        raise ValueError(errmsg)
    return sorted(set(result))
```

```python
def Leo_Francisco_Simpao(data):
    stripped_data = [x.strip() for x in data.split(",")]
    ordered_list = []
    for i in stripped_data:
        if "-" in i:
            ls = list(map(int, i.split("-")))
            for x in range(ls[0], ls[-1] + 1):
                ordered_list.append(x)
        else:
            ordered_list.append(int(i))
    return ordered_list
```

```python
def Ben_Cleary(data):
    import re

    class InvalidDigit(Exception):
        pass

    class OrderedSequenceGenerator:
        processed = []
        cleaned_payload = []

        def __init__(self, data):
            self.safety_check_and_process(data)

        def safety_check_and_process(self, data):
            pattern = re.compile("^[0-9]{1,}?(-[0-9]{1,})?$")
            for item in list(map(lambda item: item.strip(), data.split(','))):
                if pattern.match(item):
                    self.cleaned_payload.append(item)
                else:
                    raise InvalidDigit("Data is not in correct format, please use XXX-XXX // X, where X is a number")

        def order(self):
            for item in self.cleaned_payload:
                if "-" in item:
                    r = list(map(int, item.split('-')))
                    [self.processed.append(x) for x in range(r[0], r[1] + 1)]
                else:
                    self.processed.append(int(item))


    sequence = OrderedSequenceGenerator(data)
    sequence.order()
    return sequence.processed
```

```python
def Abhijit_Mukherjee(input_sequence):
    list_of_num = []
    for input in input_sequence.split(','):
        if len(input) > 1:
            for num in range(int(input.split('-')[0]),int(input.split('-')[1])+1):
                list_of_num.append(num)
        else:
            list_of_num.append(int(input))

    return list_of_num
```

```python
def Carlos_Guillermo_Durazo(lst):
  final_list = []
  for i in lst.split(','):
    if len(i.split('-'))==1:
        final_list.append(int(i))
    else:
        for j in range(
            int(i.split('-')[0]), 
            int(i.split('-')[-1])+1):
                final_list.append(j)
  return final_list
```

```python
def test_funcs(*funcs):
    import traceback
    tests = ['1-5,7,9,10-13',
             '5, 10-6, 4-8,3',
             '1, 3 -5, 6-, -, 7, -9 , 16 - 20,23',
             '2, 4, num, ;5, a-b']
    for func in funcs:
        print(f'\n\nTesting function: {func.__name__}\n')
        print('-----------')
        for test in tests:
            print(f'Testing: {test}')
            try:
                result = func(test)
            except ValueError as errmsg:
                print(errmsg)
            except Exception as e:
                print(f'ERROR: {traceback.format_exc()}')
            else:
                print(result)
            print('-----------')
```
