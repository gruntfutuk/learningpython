```python
        import copy

        def replacing_list(lst, to_replace, rep, in_place=True):
            if not in_place:
                lst = copy.deepcopy(lst)
            for idx, entry in enumerate(lst):
                if entry == to_replace:
                    lst[idx] = rep
                elif isinstance(entry, list):
                    replacing_list(entry, to_replace, rep)
            if not in_place:
                return lst


        tests = [([10, 20, 30], 20, 55),
                 (['Mary', 'had', 'a', 'little', 'lamb'], 'lamb', 'dog'),
                 ([['Mary', 'Wendy', 'Mary'], 67, 'Mary', 49, 'Mary', 'Helen', 21], 'Mary', 'Mike'),
                 ([1, 2, 3, [10, 20, 40], 4, 5], [10, 20, 40], {'name': 'me', 'age': 10})]

        for list_, target, replacement in tests:
            print('Returned:', replacing_list(list_, target, replacement, in_place=False))
            print('Original:', list_)
            replacing_list(list_, target, replacement)
            print('Replaced:', list_)
            print()
```

```python
#!/usr/bin/python3.6

def M_1():
    print('\n\nYou chose option 1.\nThat gives you another menu:')
    return 'M_1'

def M_2():
    print('\n\nYou chose option 2.\nDoes some stuff and returns you to same menu.')
    
def M_3():
    print('\n\nYou chose option 3.\nDoes some other stuff and returns you to same menu.')
    
def M1_a():
    print('\n\nNow you get a sub-menu full of more exciting options:')
    return 'M1'

def M1_b():
    print('\n\nYou chose option B of sub-menu.\nDoes some stuff and returns you to same menu.')

def M1_print():
    print('\n\nYour choice would do something exciting like print a report.')
    print('Then take you back to the same menu. Bye.')
    
def menu_choice(prompt, choices, case_sensitive):
    while True:
        print(prompt, end='')
        response = input() if case_sensitive else input().lower()
        if response and response in choices.keys():
            return response
        print('Not a valid choice. Please try again.')
    
def menu_present(menu_level):
    menu = menus.get(menu_level)
    option = menu_choice(menu['prompt'], menu['choices'], menu['case_sensitive'])
    action = menu['choices'].get(option)
    return option, action
    
menus = {'Master': {
            'choices': {'1': M_1, '2': M_2, '3': M_3, 'q': None},
            'prompt': "\n\nMENU\n1. One\n2. Two\n3. Three\nq. Quit\nChoice? ",
            'case_sensitive': False,
            'parent': 'Master'
            },
        'M_1': {
            'choices': {'a': M1_a, 'b': M1_b, 'u': None, 'q': None},
            'prompt': "\n\nSub-Menu 1\nA. Alpha\nB. Beta\nu. Up\nq. Quit\nChoice? ",
            'case_sensitive': False,
            'parent': 'Master'
            },
        'M1': {
            'choices': {'p': M1_print, 'u': None},
            'prompt': "\n\nSub-sub-menu XXX\nP. Print\nu. Up\nChoice? ",
            'case_sensitive': False,
            'parent': 'M_1'
            }
        }

menu_level = "Master"

while True:
    option, action = menu_present(menu_level)
    if not action:
         if option.lower() == 'q':
             break
         if option.lower() == 'u':
             menu_level = menus[menu_level]['parent']
    else:
        outcome = action()
        if outcome:
            menu_level = outcome
```

```python
num1 = float(input("Enter first number: "))
op = input("Enter  operator: ")
num2 = float(input("Enter second number: "))
 
if op == "+":
    print(num1 + num2)
elif op == "-":
    print(num1 - num2)
elif op == "/":
    print(num1 / num2)
elif op == "*":
    print(num1 * num2)
else:
    print("Invalid operator")
```

```python
'adds some text to the beginning of a file, and moves the'
'rest of the content down the file'

new_text = 'This is new text with a ðŸ˜® to make you laugh'
buffer = new_text.encode('utf-8')
size = len(buffer)

with open('buffer.txt', 'r+b') as f:
    while buffer:
        next_buffer = f.read(size)
        f.seek(-len(next_buffer), 1)
        f.write(buffer)
        buffer = next_buffer
```
