```python
LABELS = {int: "integer", float: "floating point number", str: "string"}

def get_input(msg, type_req = int):
    ''' prompt user for input with msg and return appropriately cast value '''
    
    while True:
        response = input(msg)
        if response:
            label = LABELS.get(type_req, "Uknown type requested")
            try:
                if type_req == int:
                    value = int(response)
                elif type_req == float:
                    value = float(response)
                elif type_req == str:
                    value = str(response)
                else:
                    raise TypeError(label)
            except ValueError as e:
                print(f'Expected a {label} input. Please try again.')
            else:
                return value
        else:
            print('An input is required. Please try again.')


a = get_input("enter 1st number: ")
b = get_input("enter 2nd number: ")
x = a + b
y = a - b
z = a * b
d = a / b
print("a + b = ", x)
print("a - b = ", y)
print("a x b = ", z)
print("a / b = ", d)
```

```python
def get_input(msg: str = '', typ: type = str, default = None, null_ok: bool = False):
    
    if typ not in [str, int, float]:
        raise TypeError(f'Type {typ} is not available in this function')
    if default and type(default) != typ:
        raise TypeError(f'Requested type, {typ} and type of default, {type(default)} do not match')
    while True:
        response = input(msg)
        if not response and (null_ok or default):
            if default:
                response = default
            else:
                response = None
            break
        elif response and typ == str:
            break
        elif response and typ == int:
            if response.isdigit():
                response = int(response)
                break
            else:
                print(f'{response} is not a valid whole number')
        elif response and typ == float:
            try:
                response = float(response)
            except ValueError as e:
                print(f'{response} is not a valid floating point number')
            else:
                break
        else:
            print('Input required')
    return response
```

```python
AFFIRMATION = frozenset(['yes', 'yeh', 'yeah', 'si', 'ok', 'sure', 'y', '1','go'])
REJECTION = frozenset(['no', 'nah', 'nope', 'n', '0'])

def yes(msg = "Yes/No? "):
    while True:
        ans = input(msg).strip().lower()
        if ans in AFFIRMATION:
            return True
        if ans in REJECTION:
            return False
        print(f'Sorry, don\'t know what {ans} means. Try again.')
    
if yes('Do you want to learn python? '):
    print('You\'ve made a great decision.')
else:
    print('Hope you change your mind.')
```
